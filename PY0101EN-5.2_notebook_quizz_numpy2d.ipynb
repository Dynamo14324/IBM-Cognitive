2)  plot the function

```python
import matplotlib.pyplot as plt
Ap=np.dot(A,B)
Bp=np.dot(B,A)
#print(Ap==Bp)

def Plotvec2(a,b):
    ax = plt.axes()
    ax.arrow(0, 0, *a, head_width=0.05, color ='r', head_length=0.1)
    plt.text(*(a + 0.1), 'a')
    ax.arrow(0, 0, *b, head_width=0.05, color ='b', head_length=0.1)
    plt.text(*(b + 0.1), 'b')
    plt.ylim(min(a), max(1))
    plt.xlim(min(b), max(b))
    
```

```python
Plotvec2(Ap[0,],Ap[1,])
```
